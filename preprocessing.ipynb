### Data Encoding

#### Using integer values instead of the actual words so that our machine learning model could interpret it.

### Encoding Gender

```python
def gender(x):
    if x == 'Female':
        return int(1)
    else:
        return int(0)
df["Gender"]=df["Gender"].apply(gender)
gender=df['Gender'].value_counts().plot(kind="bar",color=["salmon","lightblue"],title="Count of different gender")
df.iloc[:,10 : 13]
```

### Encoding Educational Field

```python
educational_field_dict = df["EducationField"].value_counts()
print(educational_field_dict)
educational_field_dict_new={
'Life Sciences':5,
'Medical':4,
'Marketing': 3,
'Technical Degree':2,
'Other':1,
'Human Resources': 0
}
print(educational_field_dict_new)

def education(x):
    if str(x) in educational_field_dict_new.keys():
        return educational_field_dict_new[str(x)]

    
df["EducationField"] = df["EducationField"].apply(education)
df.head()
```

### Encoding Marital Status

```python
def married_status(x):
    if x == 'Single':
        return int(0)
    elif x=='Divorced':
        return int(1)
    else :
        return int(2)
print(df['MaritalStatus'].value_counts())
df["MaritalStatus"]=df["MaritalStatus"].apply(married_status)
x=df['MaritalStatus'].value_counts().plot(kind="bar",color=["salmon","lightblue","teal"],title="Marital Status")
```

### Encoding Business Travel

```python
def BusinessTravel(x):
    if x == 'Travel_Rarely':
        return int(0)
    elif x=='Travel_Frequently':
        return int(1)
    else :
        return int(2)
print(df['BusinessTravel'].value_counts())
df["BusinessTravel"]=df["BusinessTravel"].apply(BusinessTravel)
x=df['BusinessTravel'].value_counts().plot(kind="bar",color=["salmon","lightblue","teal"],title="Count of Business Travel")

```

### Encoding Job Role

```python
JobRole_new={
'Sales Executive':8,
'Research Scientist':7,
'Laboratory Technician': 6,
'Manufacturing Director':5,
'Healthcare Representative':4,
'Manager': 3,
'Sales Representative':2,
'Research Director':1,
'Human Resources':0
}
print(JobRole_new)

def JobRole(x):
    if str(x) in JobRole_new.keys():
        return JobRole_new[str(x)]

    
df["JobRole"] = df["JobRole"].apply(JobRole)
x=df['JobRole'].value_counts().plot(kind="bar",title="Count of different Job Roles")

```

### Encoding Department

```python
def Department(x):
    if x == 'Research & Development':
        return int(0)
    elif x=='Sales':
        return int(1)
    else :
        return int(2)
print(df['Department'].value_counts())
df["Department"]=df["Department"].apply(Department)
x=df['Department'].value_counts().plot(kind="bar",color=["salmon","lightblue","teal"],title="Count of Business Travel")

```

### Encoding Department

```python
def OverTime(x):
    if x == 'Yes':
        return int(1)
    else:
        return int(0)
print(df['OverTime'].value_counts())
df["OverTime"]=df["OverTime"].apply(OverTime)
x=df['OverTime'].value_counts().plot(kind="bar",color=["salmon","lightblue"],title="Count of different gender")
df.iloc[:,10 : 13]
```
